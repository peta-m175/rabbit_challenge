# 勾配降下法

## 勾配降下法

深層学習の目的
  - 学習を通して誤差を最小にするネットワークを作成すること
    - 誤差$E(w)$を最小化するパラメータ$w$を発見すること
      - パラメータを最適化 -> 勾配降下法を利用

$$
  w^{(t+1)}=w^{(t)}-\varepsilon\nabla E　\\
  \nabla E = \frac{\partial E}{\partial w}
  = [\frac{\partial E}{\partial w_1}\dots \frac{\partial E}{\partial w_M}]
$$
$\varepsilon$: 学習率
- 学習率の値によって学習の効率が大きく異なる
  - 学習率が大きすぎた場合
    - 最小値にいつまでもたどり着かず発散してしまう
  - 学習率が小さい場合
    - 発散することはないが、小さすぎると収束するまでに時間がかかってしまう
    image_gradient_descent.png

勾配降下法の学習率の決定、収束性向上のためのアルゴリズムについて複数の論文が公開され、以下がよく利用されている
- Momentum
- AdaGrad
- Adadelta
- Adam

## 確率的勾配降下法(SGD)

ランダムに抽出したサンプルの誤差
$$
w^{(t+1)}=w^{(t)}-\varepsilon\nabla E_t
$$
- メリット
  - データが冗⻑な場合の計算コストの軽減
  - 望まない局所極小解に収束するリスクの軽減
  - オンライン学習ができる

## ミニバッチ勾配降下法

ランダムに分割したデータの集合（ミニバッチ）$D_t$に属するサンプルの平均誤差

$$
w^{(t+1)}=w^{(t)}-\varepsilon\nabla E_t \\
E_t = \frac{1}{N_t}\sum_{n\in D_t}E_n \\
N_t = |D_t|
$$

- メリット
  - 確率的勾配降下法のメリットを損なわず、計算機の計算資源を有効利用できる
    - CPUを利用したスレッド並列化やGPUを利用したSIMD並列化
