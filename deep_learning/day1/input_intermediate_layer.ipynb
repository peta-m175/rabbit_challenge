# 入力層~中間層

## 全体像

- 事前に用意する情報
  - 入力: $x_n=[x_{n1}\dots x_{ni}]$
  - 訓練データ: $d_n=[d_{n1}\dots d_{nk}]$
- 多層ネットワークのパラメータ
$$
w^{(l)} \begin{cases}
    重み:~~~~~~~~W^{(l)} = \left(
    \begin{array}{ccc}
      w_{11}^{(l)} & \ldots & w_{1l}^{(l)} \\
      \vdots & \ddots & \vdots \\
      w_{j1}^{(l)} & \ldots & w_{jl}^{(l)}
    \end{array}
  \right)\\
    バイアス:~~b^{(l)}=[b_1^{(l)}\dots b_j^{(l)}]
  \end{cases}
$$
- 活性化関数:
$$
f^{(l)}(u^{(l)})=[f^{(l)(u_1^{(l)})}\dots f^{(l)}(u_j^{(l)})]
$$
- 中間層出力:
$$
z^{(l)}=[z_1^{(t)}\dots z_K^{(t)}]=f^{(l)}(u^{(l)})
$$
- 総入力:
$$
u^{(l)}=W^{(l)}z^{(l-1)}+b^{(l)}
$$
- 出力:
$$
y_n^{(t)}=[y_{n1}^{(t)}\dots y_{nK}^{(t)}]=z^{(L)}
$$
- 誤差関数
$$
E_n(w)
$$
- 各インデックス
  - 入力層
  $$
  i(=1\dots I)
  $$
  - 中間層
  $$
  j(=1\dots J)
  $$
  - 出力層
  $$
  k(=1\dots K)
  $$
  - 層
  $$
  l(=1\dots L)
  $$
  - 訓練データ
  $$
  n(=1\dots N)
  $$
  - 試行回数
  $$
  t(=1\dots T)
  $$

誤差関数をもとにネットワークを学習させるには
$$
w^{(t+1)}=w^{(t)}-\varepsilon\nabla E_n(w)
$$
