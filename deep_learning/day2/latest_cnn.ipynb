# 最新のCNN

## AlexNet

- AlexNetとは2012年に開かれた画像認識コンペティション2位に大差をつけて優勝したモデル
- AlexNetの登場で、ディープラーニングが大きく注目を集めた

- モデルの構造\
5層の畳み込み層およびプーリング層など、それに続く3層の全結合層から構成される。
- 過学習を防ぐ施策
  - サイズ4096の全結合層の出力にドロップアウトを使用している
