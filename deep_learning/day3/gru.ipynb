# GRU

従来のLSTMでは、パラメータが多数存在していたため、計算負荷が大きい。\
GRUでは、そのパラメータを大幅に削減し、精度は同等またはそれ以上が望める様になった構造。


メリット
- 計算負荷が低い

## 確認テスト

### P.88

- Q: LSTMとCECが抱える課題について、それぞれ簡潔に述べよ。
- A:
  - LSTM:パラメータが多数存在するため、計算負荷が大きい
  - CEC: 時間依存度に関係なく重みが一律で、ニューラルネットワークの学習特性が無い

### P.92

- Q: LSTMとGRUの違いを簡潔に述べよ
- A: パラメータの数が異なる(LSTM:多, GRU:少)
