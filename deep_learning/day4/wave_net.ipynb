# WaveNet

- Aäron van den Oordらにより提案\
  https://arxiv.org/abs/1609.03499
- 生の音声波形を生成する深層学習モデル
- Pixel CNN(高解像度の画像を精密に生成できる手法)を音声に応用したもの

- 時系列データに対して畳み込み(**Dilated causal convolution**)を適応
  - 層が深くなるにつれて畳み込むリンクを離す
  - パラメータ数に対する受容野が広い
