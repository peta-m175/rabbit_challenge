# プルーニング

ネットワークが大きくなると大量のパラメータが全てのニューロンの計算が精度に寄与しているわけではない

-> モデルの制度に寄与が少ないニューロンを削除することでモデルの軽量化、工作科が見込まれる

## 計算の高速化

寄与が少ないニューロンの削除、モデルの圧縮を行い、高速に計算できる

## ニューロンの削減

ニューロンの削減の手法は重みが閾値以下の場合ニューロンを削減し、再学習を行う

Screenshot from 2020-12-31 10-12-53.png

## ニューロン数と精度

閾値を高くするとニューロンは大きく削減できるが精度も減少
