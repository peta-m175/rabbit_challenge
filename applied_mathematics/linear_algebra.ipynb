### 連立1次方程式


$$
x_1 + 2x_2 = 3\\
2x_1 + 5x_2 = 5
$$

上記の関係式があるとした場合、以下の様なグラフとして表すことができる。

```
%matplotlib inline

from sympy import Symbol
from sympy.plotting import plot_implicit

x = Symbol('x')  
y = Symbol('y')
# x_1 + 2x_2 = 3
equation1 = x + 2 * y - 3
# 2x_1 + 5x_2 = -1
equation2 = 2 * x + 5 * y - 5

plot1 = plot_implicit(equation1, x_var=(x, -6, 6), y_var=(y, -6, 6), line_color="blue", show=False)
plot2 = plot_implicit(equation2, x_var=(x, -6, 6), y_var=(y, -6, 6), line_color="red", show=False)

plot1.extend(plot2)
plot1.show()
```

## 固有値

### 固有値と固有ベクトル

ある行列Aに対して，以下のような式が成り立つような，特殊なベクトル$\vec{x}$と，右辺の係数$\lambda$がある。
$$
A\vec{x} = \lambda\vec{x}
$$
行列Aとその特殊なベクトル$\vec{x}$の積は、ただのスカラーの数$\lambda$とその特殊なベクトル$\vec{x}$との積が同じ値になる。 

この特殊なベクトル$\vec{x}$とその係数$\lambda$を、
行列Aに対する、固有ベクトル、固有値という。

問題
$$
\left ( \begin{array}{ccc}
  3 & 2 & 0 \\
  0 & 2 & 0 \\
  0 & 0 & 1 \\
\end{array} \right)の固有値, 固有ベクトルを求めよ
$$

$$
A \vec{x} = \lambda \vec{x} \\
(A=\lambda I)\vec{x}=\vec{0} \\
\vec{x} \neq \vec{0} より \\
\left | \begin{array}{ccc}
  3-\lambda & 2 & 0 \\
  0 & 2-\lambda & 0 \\
  0 & 0 & 1-\lambda \\
\end{array} \right| = 0 \\
(3-\lambda)(2-\lambda)(1-\lambda) = 0 \\
\lambda = 3 or 2 or 1
$$
各固有値を当てはめる。
1. if $\lambda=3$
$$
\left ( \begin{array}{ccc}
  3 & 2 & 0 \\
  0 & 2 & 0 \\
  0 & 0 & 1 \\
\end{array} \right)
\left ( \begin{array}{c}
  x_1 \\
  x_2 \\
  x_3 \\
\end{array} \right)
= 3 \left ( \begin{array}{c}
  x_1 \\
  x_2 \\
  x_3 \\
\end{array} \right)
よってx_2=0, x_3=3
$$
2. if $\lambda=2$
$$
\left ( \begin{array}{ccc}
  3 & 2 & 0 \\
  0 & 2 & 0 \\
  0 & 0 & 1 \\
\end{array} \right)
\left ( \begin{array}{c}
  x_1 \\
  x_2 \\
  x_3 \\
\end{array} \right)
= 2 \left ( \begin{array}{c}
  x_1 \\
  x_2 \\
  x_3 \\
\end{array} \right)
よってx_1=-2x_2, x_3=0
$$
3. if $\lambda=1$
$$
\left ( \begin{array}{ccc}
  3 & 2 & 0 \\
  0 & 2 & 0 \\
  0 & 0 & 1 \\
\end{array} \right)
\left ( \begin{array}{c}
  x_1 \\
  x_2 \\
  x_3 \\
\end{array} \right)
= 1 \left ( \begin{array}{c}
  x_1 \\
  x_2 \\
  x_3 \\
\end{array} \right)
よってx_1=0, x_2=0
$$

したがって
$$
\lambda=3のとき\vec{x}=
\left ( \begin{array}{c}
  1 \\
  0 \\
  0 \\
\end{array} \right)の定数倍 \\
\lambda=2のとき\vec{x}=
\left ( \begin{array}{c}
  2 \\
  -1 \\
  0 \\
\end{array} \right)の定数倍 \\
\lambda=3のとき\vec{x}=
\left ( \begin{array}{c}
  0 \\
  0 \\
  1 \\
\end{array} \right)の定数倍 \\
$$

### 固有値分解

ある実数を正方形にならべて作られた行列Aが固有値$\lambda_1,\lambda_2,\cdots$と固有ベクトル$\vec{v_1},\vec{v_2},\cdots$を持ったとする。
この固有値を対角線上に並べた行列（それ以外の成分は0）

$$
A = 
\left ( \begin{array}{ccc}
  \lambda_1 &  & \\
   & \lambda_2 & \\
   & & \ddots \\
\end{array} \right)
$$
と，それに対応する固有ベクトルを並べた行列
$$
V = 
\left ( \begin{array}{ccc}
  \vec{v_1} & \vec{v_2} & \cdots \\
\end{array} \right)
$$
を用意したとき、それらは
$$
AV=VA
$$
と関係付けられる。したがって、
$$
A=VAV^{-1}
$$
と変形できる。このように正方形の行列を上述の様な３つの行列の積に変換することを固有値分解という。この変換によって行列の累乗の計算が容易になる等の利点がある。

問題
$$
\left ( \begin{array}{cc}
  2 & 1 \\
  0 & 6 \\
\end{array} \right)を固有値分解せよ
$$

$$
A \vec{x} = \lambda \vec{x} \\
(A=\lambda I)\vec{x}=\vec{0} \\
\vec{x} \neq \vec{0} より \\
\left | \begin{array}{cc}
  2-\lambda & 1 \\
  0 & 6-\lambda \\
\end{array} \right| = 0 \\
\lambda = 2 or 6 \\
\left ( \begin{array}{cc}
  2 & 1 \\
  0 & 6 \\
\end{array} \right)
\left ( \begin{array}{c}
  x_1 \\
  x_2 \\
\end{array} \right)
= 2 \left ( \begin{array}{c}
  x_1 \\
  x_2 \\
\end{array} \right) よってx_2=0 \\
\left ( \begin{array}{cc}
  2 & 1 \\
  0 & 6 \\
\end{array} \right)
\left ( \begin{array}{c}
  x_1 \\
  x_2 \\
\end{array} \right)
= 6 \left ( \begin{array}{c}
  x_1 \\
  x_2 \\
\end{array} \right) よって4x_1=x_2
$$
したがって
$$
\lambda=2のとき\vec{x}=
\left ( \begin{array}{c}
  1 \\
  0 \\
\end{array} \right)の定数倍 \\
\lambda=6のとき\vec{x}=
\left ( \begin{array}{c}
  1 \\
  4 \\
\end{array} \right)の定数倍
$$
つまり
$$
\left ( \begin{array}{cc}
  2 & 1 \\
  0 & 6 \\
\end{array} \right)
= 
\left ( \begin{array}{cc}
  1 & 1 \\
  0 & 4 \\
\end{array} \right)
\left ( \begin{array}{cc}
  2 & 0 \\
  0 & 6 \\
\end{array} \right)
\left ( \begin{array}{cc}
  1 & 1 \\
  0 & 4 \\
\end{array} \right)^{-1} \\
\left ( \begin{array}{cc}
  2 & 1 \\
  0 & 6 \\
\end{array} \right)
= 
\left ( \begin{array}{cc}
  1 & 1 \\
  0 & 4 \\
\end{array} \right)
\left ( \begin{array}{cc}
  2 & 0 \\
  0 & 6 \\
\end{array} \right)
\left ( \begin{array}{cc}
  1 & -1/4 \\
  0 & 1/4 \\
\end{array} \right)
$$

### 特異値分解

正方行列以外の固有値分解も似たようなことができる。
$$
M\vec{v}=\sigma\vec{u} \\
M^{\tau}\vec{u}=\sigma\vec{v}
$$
このような特殊な単位ベクトルがあるならば特異値分解できる。
$$
M = USV^{-1}
$$

特異値の求め方
$$
MV = US, M^{\tau}U=VS^{\tau} \\
M =USV^{-1}, M^{\tau} = VS^{\tau}U^{-1} \\
MM^{\tau} = USV^{-1}VS^{\tau}U^{-1}=USS^{\tau}U^{-1}
$$

つまり$MM^{\tau}$を固有値分解すれば、左特異ベクトルとの特異値の2上が求められることがわかる。

（左特異ベクトルは単位ベクトルから作らなければならない）

$$
M =
\left ( \begin{array}{ccc}
  1 & 2 & 3 \\
  3 & 2 & 1 \\
\end{array} \right)\\
MM^{\tau} =
\left ( \begin{array}{ccc}
  1 & 2 & 3 \\
  3 & 2 & 1 \\
\end{array} \right)
\left ( \begin{array}{cc}
  1 & 3 \\
  2 & 2 \\
  3 & 1 \\
\end{array} \right) = 
\left ( \begin{array}{cc}
  14 & 10 \\
  10 & 14 \\
\end{array} \right)
$$
これを固有値分解して
$$
\left ( \begin{array}{cc}
  14 & 10 \\
  10 & 14 \\
\end{array} \right)
=
\left ( \begin{array}{cc}
  1\sqrt{2} & -1\sqrt{2} \\
  1\sqrt{2} & 1\sqrt{2} \\
\end{array} \right)
\left ( \begin{array}{cc}
  24 & 0 \\
  0 & 4 \\
\end{array} \right)
\left ( \begin{array}{cc}
  1\sqrt{2} & -1\sqrt{2} \\
  1\sqrt{2} & 1\sqrt{2} \\
\end{array} \right) ^ {-1}
$$
$$
M^{\tau}M =
\left ( \begin{array}{cc}
  1 & 3 \\
  2 & 2 \\
  3 & 1 \\
\end{array} \right)
\left ( \begin{array}{ccc}
  1 & 2 & 3 \\
  3 & 2 & 1 \\
\end{array} \right)
= \left ( \begin{array}{ccc}
  10 & 8 & 6 \\
  8 & 8 & 8 \\
  6 & 8 & 10 \\
\end{array} \right)
$$
これを固有値分解して
$$
\left ( \begin{array}{ccc}
  10 & 8 & 6 \\
  8 & 8 & 8 \\
  6 & 8 & 10 \\
\end{array} \right) =
\left ( \begin{array}{ccc}
  1\sqrt{3} & 1\sqrt{2} & 1\sqrt{6} \\
  1\sqrt{3} & 0 & -2\sqrt{6} \\
  1\sqrt{3} & -1\sqrt{2} & 1\sqrt{6} \\
\end{array} \right)
\left ( \begin{array}{ccc}
  24 & 0 & 0 \\
  0 & 4 & 0 \\
  0 & 0 & 0 \\
\end{array} \right)
\left ( \begin{array}{ccc}
  1\sqrt{3} & 1\sqrt{2} & 1\sqrt{6} \\
  1\sqrt{3} & 0 & -2\sqrt{6} \\
  1\sqrt{3} & -1\sqrt{2} & 1\sqrt{6} \\
